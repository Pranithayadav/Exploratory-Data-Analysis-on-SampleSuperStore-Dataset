```python
#Heatmap of the dataset
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(a.corr(), annot= True)
plt.show()
```

```python
sns.countplot(x=a['Segment'])
```

```python
sns.countplot(x=a['Category'])
```

```python
#regions
sns.countplot(x=a['Region'])
```

```python
#Orders of different states in the USA
a['State'].value_counts().plot(kind = 'bar',figsize=(20,15))
```

```python
#Top 50 cites with maximum orders
a['City'].value_counts().head(50).plot(kind = 'bar' , figsize=(20,15))
```

```python
#Bottom 50 cities with minimum orders
a['City'].value_counts().tail(50).plot(kind = 'bar' , figsize=(20,15))
```

```python
shipmode_profit=a.groupby('Ship Mode')['profit'].sum().reset_index().sort_vales(by='Profit',ascending=False
sns.catplot('Ship Mode', 'Profit', data=shipmode_profit, plot(kind='bar', aspect=3,height=5))
```

```python
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = a, color = 'r', label= 'Discount')
plt.legend()
```

```python
grouped=pd.DataFrame(a.groupby(['Ship Mode','Segment','Category','Sub-Category','State','Region'])['Quantity','Discount','Sales','Profit'].sum().reset_index())
grouped
```

```python
#So Now we Grouped or sum the sales ,profit,discount,quantity according to every state of region and also according to sub-categories sales
a.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```

```python
#sales vres profit
sns.jointplot(a['Sales'], a['Profit'], kind = "reg").annotate(stats.pearsonr)
plt.show()
```

```python
#qulaity vs profit
sns.jointplot(a['Quantity'], a['Profit'], kind = "reg").annotate(stats.pearsonr)
plt.show()
```

```python
#Discount vs profit
sns.jointplot(a['Discount'], a['Profit'], kind = "reg").annotate(stats.pearsonr)
plt.show()
```
